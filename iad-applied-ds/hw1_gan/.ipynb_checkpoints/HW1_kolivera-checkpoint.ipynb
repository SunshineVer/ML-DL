# Данные

Будем использовать данные телескопа MAGIC из UCI репозитория https://archive.ics.uci.edu/ml/datasets/MAGIC+Gamma+Telescope. Каждый объект в данных - параметры одного изображения кластера и метка этого кластера (фотон или адрон):


0. Length: major axis of ellipse [mm]
1. Width: minor axis of ellipse [mm]
2. Size: 10-log of sum of content of all pixels [in #phot]
3. Conc: ratio of sum of two highest pixels over fSize [ratio]
4. Conc1: ratio of highest pixel over fSize [ratio]
5. Asym: distance from highest pixel to center, projected onto major axis [mm]
6. M3Long: 3rd root of third moment along major axis [mm]
7. M3Trans: 3rd root of third moment along minor axis [mm]
8. Alpha: angle of major axis with vector to origin [deg]
9. Dist: distance from origin to center of ellipse [mm]
10. class: g,h # gamma (signal), hadron (background)

# Conditional WGAN

Мы будем использовать `Conditional WGAN`, который изображен на рисунке. В качестве условия `y` мы будем использовать метку класса: **0** - адрон, **1** - фотон. Таким образом, мы будем сообщать генератору для какой частицы нужно генерировать параметры изображения. 

<center><img src="img/cgan.png" width="800"></center>

Генератор $\hat{x} = G(z, y)$ будет принимать на вход шумовой вектор $z$ и вектор условий $y$, а выдавать будет сгенерированный (фейковый) вектор параметров $\hat{x}$. 

Дискриминатор $D(x, y)$ будет принимать на вход вектор параметров $x$ и вектор условий $y$, а возвращать будет рациональное число.

Обучать `Conditional WGAN` будем с такой функцией потерь:

$$L(G, D) = \frac{1}{n} \sum_{x_i \in X, y_i \in y} D(x_i, y_i) + -\frac{1}{n} \sum_{z_i \in Z, y_i \in y} D(G(z_i, y_i), y_i) \to \max_G \min_D$$

Реализуйте нейронную сеть для генератора со следующими слоями:
- Полносвязный слой со 100 нейронами;
- Слой батч-нормализации;
- ReLU функцию активации;
- Полносвязный слой со 100 нейронами;
- Слой батч-нормализации;
- ReLU функцию активации;
- Выходной слой.

```python
class Generator(nn.Module):
    def __init__(self, n_inputs, n_outputs):
        super(Generator, self).__init__()

        self.net = nn.Sequential(
            nn.Linear(n_inputs, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
            nn.Linear(100, n_outputs)
        )

    def forward(self, z, y):
        # шумовой вектор z
        # вектор условий y (метки классов: 0 или 1)
        zy = torch.cat((z, y), dim=1)
        return self.net(zy)
```

## Задание 4 (0.5 балл)

Реализуйте функцию для генерации новых объектов $X$ по вектору условий $y$.

```python
def generate(generator, y, latent_dim, device=DEVICE):
    prior = torch.distributions.Normal(torch.zeros(latent_dim).to(device),
                                   torch.ones(latent_dim).to(device))
    y = y.to(device)
    z = prior.sample((y.shape[0], )).to(device)
    X_fake = generator(z, y)
    return X_fake
```

```python
class Discriminator(nn.Module):
    def __init__(self, n_inputs):
        super(Discriminator, self).__init__()

        self.net = nn.Sequential(
            nn.Linear(n_inputs, 100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.ReLU(),
            nn.Linear(100, 1),
            nn.Sigmoid()
        )

    def forward(self, x, y):
        xy = torch.cat((x, y), dim=1)
        return self.net(xy)
```

Дискриминатор решает задачу бинарной классификации - получил ли он реальный объект на вход или фейковый.

- результат работы дискриминатора - уверенность модели в том, что объект принадлежит какому-то классу (число от 0 до 1, также можно интерпретировать как вероятность класса)
- функция сигмоиды преобразует входной вектор так, что все его выходные значения будут принадлежать промежутку от 0 до 1

## Задание 6 (0.5 балла)

Реализуйте функцию, которая принимала бы на вход набор сгенерированных и настоящих объектов, разбивала на train и test с учетом баланса классов (real и fake объектов), с соотношением 3 к 1, обучала модель линейной регрессии и градиентного бустинга, которые учились бы отличать настоящие объекты от фальшивых, после чего выводила бы accuracy score на отложенной выборке у обоих моделей.

Ответьте на вопрос: какое значение accuracy score нас бы удовлетворило больше всего? Почему?

```python
def other_agent_score(X_real, y_real, X_fake, y_fake, verbose=True, device=DEVICE):
    
    real = torch.cat((X_real, y_real), dim=1)
    fake = torch.cat((X_fake, y_fake), dim=1)

    X = torch.cat([real, fake], dim=0)
    y = torch.cat([torch.ones(real.shape[0]), torch.zeros(fake.shape[0])], dim=0)

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, shuffle=True)

    X_train = X_train.cpu().detach().numpy()
    X_test = X_test.cpu().detach().numpy()

    logreg = LogisticRegression()
    logreg.fit(X_train, y_train)

    boost = XGBClassifier()
    boost.fit(X_train, y_train)

    logreg_score = accuracy_score(y_test, logreg.predict(X_test))
    boosting_score = accuracy_score(y_test, boost.predict(X_test))

    if verbose:
        print('Logistic regression score: ' + str(logreg_score))
        print('Boosting score: ' + str(boosting_score))
        
    return logreg_score, boosting_score
```

По факту линейная регрессия или модель градиентного бустинга в этом сценарии решают задачу дискриминатора: определяют, настоящий объект или нет.

Наша задача - обучить качественный генератор, чтобы в результате было бы сложно отделить настоящий объект и фейковый. Если объекты правда сложно отличимы, то высокого аккураси нам не добиться. Поэтому в данном случае мы хотим иметь низкий аккураси.

Комментарии к коду: я несколько изменила структуру метода fit, используя тетрадку семинара как референс
- чередование итераций дискриминатора и генератора как в семинаре.
- построение графиков как в семинаре. Очень хотелось смотреть на лосс генератора и дискриминатора по отдельности во время обучения
- два главных эксперимента тут зафиксированы: обучение без ограничения весов дискриминатора и с ограничением. Спойлер: roc auc получился ниже (лучше для нас) у модели без ограничения весов, так что оставляю ее как основную.

```python
class Fitter(object):
    def __init__(
        self,
        generator,
        discriminator,
        batch_size = 32,
        n_epochs = 10,
        latent_dim = 1,
        lr = 0.0001,
        n_critic=5,
        device=DEVICE,
        verbose_num_iters=100
    ):

        self.generator = generator
        self.discriminator = discriminator
        self.batch_size = batch_size
        self.n_epochs = n_epochs
        self.latent_dim = latent_dim
        self.lr = lr
        # n_critic – число итераций обучения дискриминатора на одну итерацию обучения генератора
        self.gen_steps = 1
        self.discr_steps = n_critic

        self.prior = torch.distributions.Normal(
            torch.zeros(self.latent_dim).to(device), 
            torch.ones(self.latent_dim).to(device)
        )

        self.opt_gen = torch.optim.RMSprop(self.generator.parameters(), lr=self.lr)
        self.opt_discr = torch.optim.RMSprop(self.discriminator.parameters(), lr=self.lr)

        self.generator.to(DEVICE)
        self.discriminator.to(DEVICE)

        self.loss_history = []
        self.verbose_num_iters = verbose_num_iters

    def discr_loss(self, X_real, X_fake, y_cond):
        return - torch.mean(self.discriminator(X_real, y_cond)) + torch.mean(self.discriminator(X_fake, y_cond))
    
    def gen_loss(self, X_fake, y_cond):
        return - torch.mean(self.discriminator(X_fake, y_cond))

    def fit(self, X_train, y_train):

        # numpy to tensor
        X_real = torch.tensor(X_train, dtype=torch.float, device=DEVICE)
        y_cond = torch.tensor(y_train, dtype=torch.float, device=DEVICE)

        # tensor to dataset
        dataset_real = TensorDataset(X_real, y_cond)

        # Turn on training
        self.generator.train(True)
        self.discriminator.train(True)

        gen_loss_history = []
        discr_loss_history = []
        logreg_history = []
        boosting_history = []

        iter_i = 0

        # Fit GAN
        for epoch in range(self.n_epochs):
            print(f"Epoch {epoch + 1}")
            # смысл такой же как и в for batch in dataloader (итерации обучения):
            # здесь отдельно прописать обучение генератора и дискриминатора
            for (real_batch, cond_batch) in DataLoader(dataset_real, batch_size=self.batch_size, shuffle=True):  
                num_objects = real_batch.shape[0]
                z = self.prior.sample((num_objects, ))
                fake_batch = self.generator(z, cond_batch)

                if (iter_i % (self.gen_steps + self.discr_steps)) < self.gen_steps:
                    # шаг обучения генератора
                    self.opt_gen.zero_grad()
                    gen_loss = self.gen_loss(fake_batch, cond_batch)
                    gen_loss.backward()
                    self.opt_gen.step()
                    gen_loss_history.append((iter_i, gen_loss.item()))
                else:
                    # шаг обучения дискриминатора
                    self.opt_discr.zero_grad()
                    discr_loss = self.discr_loss(real_batch, fake_batch, cond_batch)
                    discr_loss.backward()
                    self.opt_discr.step()
                    # ОГРАНИЧЕНИЕ ВЕСОВ В ПОЛНОСВЯЗНЫХ СЛОЯХ ДИСКРИМИНАТОРА
                    # for layer in self.discriminator.net[::2]:
                    #     layer.weight.data.clamp_(-0.01, 0.01)
                    discr_loss_history.append((iter_i, discr_loss.item()))
                
                iter_i += 1

                # раз в verbose_num_iters визуализируем наши лоссы и семплы
                if iter_i % self.verbose_num_iters == 0:
                    clear_output(wait=True)
                    plt.figure(figsize=(20, 5))

                    plt.subplot(1, 4, 1)
                    plt.xlabel("Iteration")
                    plt.ylabel("Loss value")
                    plt.title("Generator loss")
                    plt.plot(
                        [p[0] for p in gen_loss_history], 
                        [p[1] for p in gen_loss_history]
                    )

                    plt.subplot(1, 4, 2)
                    plt.xlabel("Iteration")
                    plt.title("Discriminator loss")
                    plt.plot(
                        [p[0] for p in discr_loss_history],
                        [p[1] for p in discr_loss_history],
                        color="orange",
                    )

                    plt.subplot(1, 4, 3)
                    plt.xlabel("Epoch")
                    plt.title("Total loss")
                    plt.plot(self.loss_history, color="green")
                    
                    plt.subplot(1, 4, 4)
                    plt.xlabel("Epoch")
                    plt.title("Other agent accuracy")
                    plt.plot(
                        [p[0] for p in logreg_history],
                        [p[1] for p in logreg_history],
                        color="red", label="Logistic regression"
                    )
                    
                    plt.plot(
                        [p[0] for p in boosting_history],
                        [p[1] for p in boosting_history],
                        color="purple", label="Gradient boosting"
                    )
                    plt.legend()
                    plt.show()
                    
            # calculate and store loss after an epoch
            Z_noise = torch.normal(0, 1, (len(X_real), self.latent_dim)).to(DEVICE)
            X_fake = self.generator(Z_noise, y_cond)

            loss_epoch = torch.mean(self.discriminator(X_real, y_cond)) - torch.mean(
                self.discriminator(X_fake, y_cond)
            )
            self.loss_history.append(loss_epoch.detach().cpu())
            
            # Создайте выборку из 1000 объектов из X_train и 1000 сгенерированных объектов 
            # И запустите работу фукнции other_agent_score

            oa_loader = DataLoader(dataset_real, batch_size=1000, shuffle=True)
            oa_real_batch, oa_cond_batch = next(iter(oa_loader))
            oa_fake_batch = generate(self.generator, oa_cond_batch, self.latent_dim)
            log_reg, boost = other_agent_score(oa_real_batch, oa_cond_batch, oa_fake_batch, oa_cond_batch, verbose=False)
            logreg_history.append((epoch, log_reg))
            boosting_history.append((epoch, boost))
            
        # Turn off training
        self.generator.train(False)
        self.discriminator.train(False)
```

## Задание 10 (0.5 балла)

Реализуйте фукнцию из семинара для зашумления данных, адаптировав ее под наш тип данных.

```python
def corrupt(x, amount):
  """Corrupt the input `x` by mixing it with noise according to `amount` -> зашумляем картинку """
  noise = torch.rand_like(x)
  amount = amount.view(-1, 1)
  return x*(1-amount) + noise*amount 
```

```python
class DiffusionGenerator(nn.Module):
    def __init__(self, n_inputs, n_outputs, num_classes=2, class_emb_size=1):
        super(DiffusionGenerator, self).__init__()

        self.model = nn.Sequential(
            c(n_inputs, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
            nn.Linear(100, 250),
            nn.BatchNorm1d(250),
            nn.ReLU(),
            nn.Linear(250, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
            nn.Linear(100, n_outputs)
        )

    def forward(self, x, y):
        # входной вектор x
        # вектор условий y (метки классов: 0 или 1)
        net_input = torch.cat((x, y), dim=1)
        return self.model(net_input)
```

```python
def generate_with_diffusion(model, y, latent_dim, sheduler, device=DEVICE):
    x = torch.randn(latent_dim).to(device)

    # Sampling loop
    for i, t in enumerate(noise_scheduler.timesteps):

        # Get model pred
        with torch.no_grad():
            residual = model(x, y)

        # Update sample with step
        x = noise_scheduler.step(residual, t, x).prev_sample
    return x.detach().cpu()
```

```python
class DiffusionFitter(object):
    def __init__(
        self,
        model,
        batch_size = 32,
        n_epochs = 10,
        latent_dim = 1,
        lr = 0.0001,
        device=DEVICE
    ):

        self.model = model
        self.batch_size = batch_size
        self.n_epochs = n_epochs
        self.latent_dim = latent_dim
        self.lr = lr
        self.device = DEVICE

        self.opt = torch.optim.RMSprop(self.model.parameters(), lr=self.lr)

        self.model.to(DEVICE)
        self.loss_fn = nn.MSELoss()
        
        self.loss_history = []
        self.logreg_history = []
        self.boosting_history = []

    def fit(self, X, y):

        # numpy to tensor
        X_real = torch.tensor(X, dtype=torch.float, device=self.device)
        y_cond = torch.tensor(y, dtype=torch.float, device=self.device)

        # tensor to dataset
        dataset_real = TensorDataset(X_real, y_cond)

        # Turn on training
        self.model.train(True)

        # Fit GAN
        for epoch in range(self.n_epochs):
            loss_epoch = 0
            for i, (real_batch, cond_batch) in enumerate(
                DataLoader(dataset_real, batch_size=self.batch_size, shuffle=True)
            ):
                noise_amount = torch.rand(real_batch.shape[0]).to(self.device)
                noisy_batch = corrupt(real_batch, noise_amount)

                pred = self.model(noisy_batch, cond_batch)

                loss = self.loss_fn(pred, real_batch)

                self.opt.zero_grad()
                loss.backward()
                self.opt.step()

                noise_amount = torch.rand(X_real.shape[0]).to(self.device)
                noisy_x = corrupt(X_real, noise_amount) 

                loss_epoch += loss.cpu().detach().numpy()
            
            self.loss_history.append(loss_epoch)
            
            # Создайте выборку из 1000 объектов из X_train и 1000 сгенерированных объектов 
            # И запустите работу фукнции other_agent_score
            
            ### YOUR CODE IS HERE ######
            oa_loader = DataLoader(dataset_real, batch_size=1000, shuffle=True)
            oa_real_batch, oa_cond_batch = next(iter(oa_loader))
            oa_fake_batch = generate(self.model, oa_cond_batch, self.latent_dim)
            log_reg, boost = other_agent_score(oa_real_batch, oa_cond_batch, oa_fake_batch, oa_cond_batch, verbose=False)
            self.logreg_history.append((epoch, log_reg))
            self.boosting_history.append((epoch, boost))

            clear_output(wait=True)
            plt.figure(figsize=(20, 5))

            plt.subplot(1, 3, 1)
            plt.xlabel("Epoch")
            plt.ylabel("Loss value")
            plt.title("Diffusion model loss")
            plt.plot(self.loss_history)
            
            plt.subplot(1, 3, 2)
            plt.xlabel("Epoch")
            plt.title("Other agent loss")
            plt.plot(
                [p[0] for p in self.logreg_history],
                [p[1] for p in self.logreg_history],
                color="red", label="Logistic regression"
            )
            
            plt.plot(
                [p[0] for p in self.boosting_history],
                [p[1] for p in self.boosting_history],
                color="purple", label="Gradient boosting"
            )
            plt.legend()
            plt.show()
            ### THE END OF YOUR CODE ###
            
        # Turn off training
        self.model.train(False)
```
